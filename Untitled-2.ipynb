```python
# Overfitting check: train vs test metrics, residuals, and learning curve
from sklearn.metrics import mean_absolute_error, mean_squared_error, r2_score
from sklearn.model_selection import learning_curve
import numpy as np
import matplotlib.pyplot as plt

def find_arrays(order):
    return [globals().get(name) for name in order if globals().get(name) is not None]

def get_preds(model):
    train_cands = find_arrays(['X_train_poly','X_train_scaled','X_train'])
    test_cands  = find_arrays(['X_test_poly','X_test_scaled','X_test'])
    for Xtr in train_cands:
        for Xte in test_cands:
            try:
                return model.predict(Xtr), model.predict(Xte), Xtr, Xte
            except Exception:
                pass
    raise RuntimeError("No compatible train/test arrays found for this model.")

def metrics(y_true, y_pred):
    mse = mean_squared_error(y_true, y_pred)
    return {"MAE": mean_absolute_error(y_true, y_pred),
            "MSE": mse, "RMSE": np.sqrt(mse), "R2": r2_score(y_true, y_pred)}

def check_model(model, model_name='model'):
    ytr_pred, yte_pred, Xtr_used, Xte_used = get_preds(model)
    mtr = metrics(y_train, ytr_pred)
    mte = metrics(y_test,  yte_pred)
    print(f"{model_name} — train on `{Xtr_used.shape}`, test on `{Xte_used.shape}`")
    print(" Train:", {k: round(v,4) for k,v in mtr.items()})
    print(" Test :", {k: round(v,4) for k,v in mte.items()})
    diff_r2 = mtr['R2'] - mte['R2']
    if diff_r2 > 0.1:
        print(" → Warning: possible overfitting (train R2 much higher than test R2).")
    elif diff_r2 < -0.1:
        print(" → Possible underfitting / data leak (test R2 higher).")
    else:
        print(" → No large gap detected.")

    # Plots: Actual vs Predicted and residuals (train/test)
    fig, axs = plt.subplots(2, 2, figsize=(12,9))
    axs[0,0].scatter(y_train, ytr_pred, alpha=0.5)
    axs[0,0].plot([y_train.min(), y_train.max()],[y_train.min(), y_train.max()],'k--')
    axs[0,0].set_title('Train: Actual vs Predicted')
    axs[0,1].scatter(y_test, yte_pred, alpha=0.6, color='orange')
    axs[0,1].plot([y_test.min(), y_test.max()],[y_test.min(), y_test.max()],'k--')
    axs[0,1].set_title('Test: Actual vs Predicted')

    axs[1,0].hist(y_train - ytr_pred, bins=30, color='tab:blue', alpha=0.7)
    axs[1,0].axvline(0, color='k', linestyle='--'); axs[1,0].set_title('Train Residuals')
    axs[1,1].hist(y_test  - yte_pred, bins=30, color='orange', alpha=0.7)
    axs[1,1].axvline(0, color='k', linestyle='--'); axs[1,1].set_title('Test Residuals')
    plt.tight_layout(); plt.show()

    # Learning curve (use the first compatible feature matrix)
    X_for_lc = find_arrays(['X_train_poly','X_train_scaled','X_train'])[0]
    train_sizes, train_scores, val_scores = learning_curve(model, X_for_lc, y_train, cv=5,
                                                           train_sizes=np.linspace(0.1,1.0,5),
                                                           scoring='r2', n_jobs=-1)
    train_mean = np.mean(train_scores, axis=1)
    val_mean   = np.mean(val_scores, axis=1)
    plt.figure(figsize=(6,4))
    plt.plot(train_sizes, train_mean, 'o-', label='Train R2')
    plt.plot(train_sizes, val_mean,   'o-', label='CV R2')
    plt.xlabel('Training examples'); plt.ylabel('R2'); plt.title(f'Learning Curve ({model_name})')
    plt.legend(); plt.grid(True); plt.show()

# Example: check both models if available
if 'lr' in globals():
    check_model(lr, 'LinearRegression (lr)')
if 'gbr' in globals():
    check_model(gbr, 'GradientBoostingRegressor (gbr)')
```
